```python
def follow_rules_base(left,right):
    for no in range(len(right)):
        #if type(left[no]) == int and type(right[no]) == int:
            if no >= len(left):
                return True
            if type(left[no]) == list and type(right[no]) == int: # This is fishy, but seems to work
                return False
            if type(left[no]) == int and type(right[no]) == list: 
                return False
            if left[no] > right[no]:
                return False 
            if left[no] < right[no]:
                return True
            if no == len(right) - 1 and len(left) > len(right):
                return False
        
```

```python
def follow_rules(left,right):   
    #print(f"Checking {left} vs {right}")

    if type(left) == int and type(right) == list:
        new_left = []
        new_left.append(left)
        left = new_left
    if type(left) == list and type(right) == int:        
        new_right = []
        new_right.append(right)
        right = new_right
    if len(left) == 0 and len(right) > 0: #left runs out --> Right order
        return True
    if len(right) == 0 and len(left) > 0: #right runs out --> Wrong order
        return False
    if left[0] == [] and right[0] == []:        
        left.pop(0)
        right.pop(0)
        return follow_rules(left,right)

    if list in (type(left[0]), type(right[0])): #At least one is a list        
        subcheck = follow_rules(left[0],right[0]) #Check again with sublists
        if type(subcheck) == bool:
            return subcheck
        right.pop(0)
        left.pop(0)
        return follow_rules(left,right)
    else: #Neither side is a list  
        subcheck = follow_rules_base(left,right)
        if type(subcheck) == bool:
            return subcheck

```

```python
pair = input[125].split("\n")
left = ast.literal_eval(pair[0])
right = ast.literal_eval(pair[1])
follow_rules(left,right)
```

```python
score = 0
for pair_no in range(len(input)):
    pair = input[pair_no].split("\n")
    left = ast.literal_eval(pair[0])
    right = ast.literal_eval(pair[1])
   # print(f"Pair number {pair_no+1} ({left} and {right}) follows rules: {follow_rules(left,right)}")
   # print(pair_no)
    if follow_rules(left, right):
        score += pair_no + 1
print(f"Total score is {score}")
```

```python
def get_digit(sig, no):
    if sig == []:
        return -1
    if type(sig[0]) != int:
        return get_digit(sig[0], no)
    return sig[0]
```

```python
sig_list = []
sig_list.append([[2]])
sig_list.append([[6]])
for pair_no in range(len(input)):
    pair = input[pair_no].split("\n")
    sig_list.append(ast.literal_eval(pair[0]))
    sig_list.append(ast.literal_eval(pair[1]))
sig_list
```

```python
sig_ord = []
sig_only = []
for sig in sig_list:
    sig_add = []
    sig_add.append(get_digit(sig,0))
    sig_add.append(sig)
    sig_ord.append(sig_add)
    sig_only.append(get_digit(sig,0))
sig_only.sort()
```
