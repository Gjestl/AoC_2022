```python
#get start pos:
def get_start_pos(input):
    for i in range(len(input)):
        for j in range(len(input[i])):
            if input[i][j] == "S":
                return (i,j)
            
```

```python
#get end pos:
def get_end_pos(input):
    for i in range(len(input)):
        for j in range(len(input[i])):
            if input[i][j] == "E":
                return (i,j)
            
```

```python
def get_candidate_moves_desc(score_tracker, input, pos):
    current_height = ord(input[pos[0]][pos[1]])
    if input[pos[0]][pos[1]] == "E":
        current_height = ord("z")

    current_steps = score_tracker[pos[0]][pos[1]]
    candidate_moves = []

    if input[pos[0]][pos[1]] == "a" or input[pos[0]][pos[1]]=="b":
        if pos[0] < len(input) - 1 and input[pos[0]+1][pos[1]] == "S":
            candidate_moves.append((pos[0]+1, pos[1]))
        if pos[0] > 0 and input[pos[0]-1][pos[1]] == "S":
            candidate_moves.append((pos[0]-1, pos[1]))
        if pos[1] < len(input[pos[0]]) - 1 and input[pos[0]][pos[1] + 1] == "S":
            candidate_moves.append((pos[0], pos[1]+1))
        if pos[1] > 0 and input[pos[0]][pos[1]-1] == "S":
            candidate_moves.append((pos[0], pos[1]-1))

    if pos[0] < len(input) - 1 and abs(ord(input[pos[0]+1][pos[1]]) - current_height) < 2:
        candidate_moves.append((pos[0]+1, pos[1]))
    if pos[0] > 0 and abs(ord(input[pos[0]-1][pos[1]]) - current_height) < 2:
        candidate_moves.append((pos[0]-1, pos[1]))
    if pos[1] < len(input[pos[0]]) - 1 and abs(ord(input[pos[0]][pos[1]+1]) - current_height) < 2:
        candidate_moves.append((pos[0], pos[1]+1))
    if pos[1] > 0 and abs(ord(input[pos[0]][pos[1]-1]) - current_height) < 2:
        candidate_moves.append((pos[0], pos[1]-1))

    for move in candidate_moves:        
        if score_tracker[move[0]][move[1]] > current_steps + 1:
            score_tracker[move[0]][move[1]] = current_steps + 1

    return candidate_moves
    return score_tracker
    
    
```

```python
def get_next_pos(score_tracker, avoid_list):
    highest = -1
    high_pos = (0,0)
    for i in range(len(score_tracker)):
        for j in range(len(score_tracker[i])):
            if score_tracker[i][j] > highest and score_tracker[i][j] < 999999 and (i,j) not in avoid_list:
                high_pos = (i,j)
                highest = score_tracker[i][j]
    return high_pos
```
